```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import glob
import os
```

```python
tables = glob.glob('/content/drive/MyDrive/DDD/Merged_Dataset/*.csv')
```

```python
tables
```

```python
df_list = []
for table in tables:
    dff = pd.read_csv(table)
    dff = dff.iloc[:,2:]
    dff = dff[(dff['State'] != 5) & (dff['State'] != 1)]
    dff_alert = dff[dff['State'] == 0].copy()
    dff_drowsy = dff[dff['State'] == 10].copy()

    print("original alert",dff_alert)
    print("original drowsy",dff_drowsy)

    #z-score normalisation performed by standard scaler
    #Calculating normalised features for each row in the original dataframe
    # dff["EAR_N"] = (dff["EAR"]-dff["EAR_mean"])/ dff["EAR_std"]
    # dff["MAR_N"] = (dff["MAR"]-dff["MAR_mean"])/ dff["MAR_std"]
    # dff["CIR_N"] = (dff["CIR"]-dff["CIR_mean"])/ dff["CIR_std"]
    # dff["MOE_N"] = (dff["MOE"]-dff["MOE_mean"])/ dff["MOE_std"]

    scaler_alert = StandardScaler()

    scaler_alert.fit(dff_alert.iloc[:3,:-1])

    dff_alert[['EAR_N','MAR_N','CIR_N','MOE_N']] = scaler_alert.transform(dff_alert.iloc[:,:-1])
    dff_drowsy[['EAR_N','MAR_N','CIR_N','MOE_N']] = scaler_alert.transform(dff_drowsy.iloc[:,:-1])


    dff = pd.concat([dff_alert,dff_drowsy], axis=0, ignore_index=True)
    df_list.append(dff)

Combined = pd.concat(df_list,axis=0, ignore_index=True)
```
